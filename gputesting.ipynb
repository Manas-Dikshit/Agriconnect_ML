```python
import torch

if torch.cuda.is_available():
	device_name = torch.cuda.get_device_name(0)
	print(device_name)
	# T4   → use fp16=True,  bf16=False
	# A100 → use fp16=False, bf16=True
	# V100 → use fp16=True,  bf16=False
else:
	print("CUDA not available; using CPU.")
	# T4   → use fp16=True,  bf16=False
	# A100 → use fp16=False, bf16=True
	# V100 → use fp16=True,  bf16=False
```
